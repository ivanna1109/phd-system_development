```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diabetes_data = pd.read_csv('data/diabetes_binary_5050split_health_indicators_BRFSS2015.csv')
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(x='Diabetes_binary', data=diabetes_data, palette='viridis')
plt.title('Raspodela ciljne varijable (Diabetes_binary)')
plt.xlabel('Dijabetesa')
plt.ylabel('Broj osoba')
#plt.show()
plt.savefig('plots/1_diabetes_balance.png')
plt.close()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=diabetes_data, x='BMI', hue='Diabetes_binary', fill=True, common_norm=False, alpha=.5)
plt.title('Raspodela BMI u zavisnosti od prisustva dijabetesa')
plt.xlabel('BMI (Body Mass Index)')
plt.ylabel('Gustina')
plt.legend(title='Stanje', labels=['Dijabetes', 'Nema dijabetes'])
#plt.show()
plt.savefig('plots/2_bmi_distribution.png')
plt.close()
```

```python
risk_factors = ['HighBP', 'HighChol', 'HeartDiseaseorAttack', 'Stroke', 'DiffWalk']
labels_map = {0.0: 'Ne', 1.0: 'Da'}

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))
axes = axes.flatten()

fig.delaxes(axes[5])

fig.suptitle('Procenat pojave dijabetesa u zavisnosti od ključnih rizičnih faktora', fontsize=16)

for i, factor in enumerate(risk_factors):
    # Kreiranje tabele za procentualni prikaz
    counts = diabetes_data.groupby(factor)['Diabetes_binary'].value_counts(normalize=True).mul(100).rename('percentage').reset_index()
    
    sns.barplot(
        x=factor, 
        y='percentage', 
        hue='Diabetes_binary', 
        data=counts, 
        ax=axes[i],
        palette='magma'
    )
    
    axes[i].set_title(f'Faktor: {factor}')
    axes[i].set_xlabel(f'{factor} (0=Ne, 1=Da)')
    axes[i].set_ylabel('Procenat')
    axes[i].legend(title='Prisustvo dijabetesa')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) 
plt.savefig('plots/3_top_risk_factors.png')
plt.close()
```

```python
plt.figure(figsize=(18, 15))
corr_matrix = diabetes_data.drop('Diabetes_binary', axis=1).corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5)
plt.title('Matrica korelacija svih atributa')
plt.savefig('plots/4_correlation_heatmap.png')
plt.close()
```

```python
gen_hlth_counts = diabetes_data.groupby('GenHlth')['Diabetes_binary'].value_counts(normalize=True).mul(100).rename('percentage').reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='GenHlth', y='percentage', hue='Diabetes_binary', data=gen_hlth_counts[gen_hlth_counts['Diabetes_binary'] == 1], palette='crest')
plt.title('Procenat slučajeva dijabetesa prema kategoriji opšteg zdravlja (1=odlično, 5=loše)')
plt.xlabel('Opšte zdravlje (GenHlth)')
plt.ylabel('Procenat njih koji imaju dijabetes')
plt.xticks(ticks=range(5), labels=['Odlično', 'Vrlo dobro', 'dobro', 'okej', 'loše'])
plt.savefig('plots/5_genhlth_diabetes_correlation.png')
plt.close()
```

```python
income_counts = diabetes_data.groupby('Income')['Diabetes_binary'].value_counts(normalize=True).mul(100).rename('percentage').reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Income', y='percentage', hue='Diabetes_binary', data=income_counts[income_counts['Diabetes_binary'] == 1], palette='flare')
plt.title('Procenat slučajeva dijabetesa prema kategoriji prihoda (1=niska, 8=visoka)')
plt.xlabel('Kategorija prihoda')
plt.ylabel('Procenat njih koji imaju dijabetes')
plt.savefig('plots/6_income_diabetes_correlation.png')
plt.close()
```

```python
age_diabetes_rate = diabetes_data.groupby('Age')['Diabetes_binary'].mean() * 100

plt.figure(figsize=(12, 6))
sns.lineplot(x=age_diabetes_rate.index, y=age_diabetes_rate.values, marker='o', color='darkred') 
plt.title('Procenat obolelih od dijabetesa po starosnoj grupi (AgeGroup: 1 do 13)')
plt.xlabel('Starosna grupa (Ordinalna skala)')
plt.ylabel('Stopa dijabetesa (%)')
plt.xticks(ticks=age_diabetes_rate.index, 
           labels=[f'{i}' for i in age_diabetes_rate.index])
plt.grid(True)
plt.savefig('plots/7_age_line_trend.png')
plt.close()
```

```python
sex_labels = ['Žena', 'Muškarac']

fig, axes = plt.subplots(1, 2, figsize=(14, 7))
fig.suptitle('Distribucija pola (Sex) u zavisnosti od statusa dijabetesa', fontsize=16)

# 8a: Pie plot za grupu BEZ dijabetesa
df_no_diabetes = diabetes_data[diabetes_data['Diabetes_binary'] == 0]
sex_counts_no = df_no_diabetes['Sex'].value_counts()
axes[0].pie(sex_counts_no, labels=sex_labels, autopct='%1.1f%%', startangle=90, colors=['#A3D4A3', '#5E9F5E']) # Nijanse zelene
axes[0].set_title('Grupa: Nema dijabetes')

# 8b: Pie plot za grupu sa dijabetesom
df_with_diabetes = diabetes_data[diabetes_data['Diabetes_binary'] == 1]
sex_counts_with = df_with_diabetes['Sex'].value_counts()
axes[1].pie(sex_counts_with, labels=sex_labels, autopct='%1.1f%%', startangle=90, colors=['#F5B96B', '#E67E22']) # Nijanse narandžaste
axes[1].set_title('Grupa: Ima dijabetes')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('plots/8_sex_pie_comparison.png')
plt.close()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Status dijabetesa u zavisnosti od navika u ishrani (proporcionalno)', fontsize=16)

# 9a: Fruits (Voće)
fruits_proportions = diabetes_data.groupby('Fruits')['Diabetes_binary'].value_counts(normalize=True).mul(100).rename('percentage').reset_index()
sns.barplot(x='Fruits', y='percentage', hue='Diabetes_binary', data=fruits_proportions, ax=axes[0], palette='YlGn', dodge=False)
axes[0].set_title('Konzumacija voća (0=Ne, 1=Da)')
axes[0].set_xlabel('Jede voće?')
axes[0].set_ylabel('Procenat unutar kategorije (%)')
axes[0].set_xticks(ticks=[0, 1], labels=['Ne', 'Da'])
axes[0].legend(title='Dijabetes')

# 9b: Veggies (Povrće)
veggies_proportions = diabetes_data.groupby('Veggies')['Diabetes_binary'].value_counts(normalize=True).mul(100).rename('percentage').reset_index()
sns.barplot(x='Veggies', y='percentage', hue='Diabetes_binary', data=veggies_proportions, ax=axes[1], palette='Oranges', dodge=False)
axes[1].set_title('Konzumacija povrća (0=Ne, 1=Da)')
axes[1].set_xlabel('Jede povrće?')
axes[1].set_ylabel('Procenat unutar kategorije (%)')
axes[1].set_xticks(ticks=[0, 1], labels=['Ne', 'Da'])
axes[1].legend(title='Dijabetes')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('plots/9_diet_stacked_comparison.png')
plt.close()
```
